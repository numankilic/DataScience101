### Soru 1:
* Stackoverflow'un 2019 yılı için geliştiricilerle yaptığı anket sonuçlarını içeren ```"survey_results_public.csv"``` adlı dosyayı okuyarak __survey_19__ adlı DataFrame'e aktarın.
* DataFrame'in satır ve sütun sayısını yazdırın.
* ```Respondent``` sütununu index olarak atayın.
* Aşağıdaki sütunları dataframe'den siliniz:

    ``` ['ScreenName', 'SOVisit1st','SOVisitFreq', 'SOVisitTo', 'SOFindAnswer', 'SOTimeSaved', 'SOHowMuchTime', 'SOAccount',   'SOPartFreq', 'SOJobs', 'EntTeams', 'SOComm', 'WelcomeChange', 'SONewContent']``` 
  
  
* ```DevType``` ya da ```LanguageWorkedWith``` sütununda eksik veri olan satırları siliniz. Kaç satır silindi?
* ```Country``` sütunundaki ```Russian Federation``` ve ```Czech Republic``` değerlerini sırasıyla ```Russia``` ve ```Czechia``` olarak değiştiriniz.
* Haftalık çalışma saati ```(WorkWeekHrs)``` ya da yıllık maaş ```(ConvertedComp)``` sütunlarındaki değerleri, ilgili sütunun ortalamasının %99'undan daha uzakta veri içeren satırları siliniz.
* ```Gender``` sütununda ```Man```, ```Woman``` ve ```NaN``` dışındaki değerleri, ```Non-binary``` değeri ile değiştirin.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
#Stackoverflow'un 2019 yılı için geliştiricilerle yaptığı anket sonuçlarını içeren "survey_results_public.csv"
#adlı dosyayı okuyarak survey_19 adlı DataFrame'e aktarın.
survey_19 = pd.read_csv(".\\developer_survey_2019\\survey_results_public.csv")

#DataFrame'in satır ve sütun sayısını yazdırın.
count_row = survey_19.shape[0]
count_column = survey_19.shape[1]
print("Satır sayısı= ",count_row," Sütun sayısı= " ,count_column)

#Respondent sütununu index olarak atayın.
survey_19.set_index('Respondent', inplace = True)

#Aşağıdaki sütunları dataframe'den siliniz:
#['ScreenName', 'SOVisit1st','SOVisitFreq', 'SOVisitTo', 'SOFindAnswer', 'SOTimeSaved', 'SOHowMuchTime', 'SOAccount',
#'SOPartFreq', 'SOJobs', 'EntTeams', 'SOComm', 'WelcomeChange', 'SONewContent']
survey_19.drop(['ScreenName', 'SOVisit1st','SOVisitFreq', 'SOVisitTo',
                'SOFindAnswer', 'SOTimeSaved', 'SOHowMuchTime', 'SOAccount',
                'SOPartFreq', 'SOJobs', 'EntTeams', 'SOComm', 'WelcomeChange', 'SONewContent'], axis=1, inplace = True)

#DevType ya da LanguageWorkedWith sütununda eksik veri olan satırları siliniz. Kaç satır silindi?
survey_19.dropna(subset=['DevType', 'LanguageWorkedWith'], inplace = True)
count_row2= survey_19.shape[0]
count_deleted_row = count_row-count_row2
print("Silinen satır sayısı: ",count_deleted_row)

#Country sütunundaki Russian Federation ve Czech Republic değerlerini sırasıyla Russia ve Czechia olarak değiştiriniz.
survey_19['Country'].replace({"Russian Federation":"Russia","Czech Republic":"Czechia"}, inplace=True)

#Haftalık çalışma saati (WorkWeekHrs) ya da yıllık maaş (ConvertedComp) sütunlarındaki değerleri, 
#ilgili sütunun ortalamasının %99'undan daha uzakta veri içeren satırları siliniz.
survey_19.drop(survey_19[(survey_19['WorkWeekHrs'] > survey_19['WorkWeekHrs'].mean()*1.99)
                         |(survey_19['WorkWeekHrs'] < survey_19['WorkWeekHrs'].mean()*0.01)].index, inplace = True)

survey_19.drop(survey_19[(survey_19['ConvertedComp'] > survey_19['ConvertedComp'].mean()*1.99)|
                         (survey_19['ConvertedComp'] < survey_19['ConvertedComp'].mean()*0.01)].index, inplace = True)

#Gender sütununda Man, Woman ve NaN dışındaki değerleri, Non-binary değeri ile değiştirin.
words1=['Man', 'Woman', np.nan]
survey_19.loc[~survey_19['Gender'].isin(words1), 'Gender'] = 'Non-binary'
````

### Soru 2:
* ```Age``` sütununu 0, 15, 19, 24, 29, 34, 39, 44, 49, 54, 59, 99 binleri ile aralığa çevirerek ```AgeInterval``` adında yeni bir sütun oluşturunuz.
* Ankete katılanların yaş aralıklarına göre dağılımını pasta grafiğinde gösteriniz. 
* (19-24] yaş aralığının en çok kullandığı sosyal medya nedir?

````python
#Age sütununu 0, 15, 19, 24, 29, 34, 39, 44, 49, 54, 59, 99 binleri ile aralığa çevirerek
#AgeInterval adında yeni bir sütun oluşturunuz.
arr2 = [0, 15, 19, 24, 29, 34, 39, 44, 49, 54, 59, 99]
survey_19['AgeInterval'] = pd.cut(x=survey_19['Age'], bins = arr2)

#Ankete katılanların yaş aralıklarına göre dağılımını pasta grafiğinde gösteriniz.
AgeInterval = survey_19['AgeInterval'].value_counts().index
counts = survey_19['AgeInterval'].value_counts().values
plt.pie(counts,labels=AgeInterval, radius=3, autopct='%1.2f%%')
plt.show()

#(19-24] yaş aralığının en çok kullandığı sosyal medya nedir?
max_social_media_use = survey_19['SocialMedia'].loc[(survey_19['Age']<=24)|(survey_19['Age']>19)].value_counts().idxmax()
print("(19-24] yaş aralığının en çok kullandığı sosyal medya "+ max_social_media_use+"dir.")
````

### Soru 4: 
Kadınların daha çok tercih ettiği/edildiği (ankete katılan kadın-erkek oranından daha yüksek orana sahip) geliştirici tipleri ```(DevType)```nelerdir?
* Her bir ```DevType``` tipinin sütun olarak, ankete katılanların da satır olarak temsil edildiği yeni bir DataFrame oluşturup __devtype_df__ adlı değişkene atayın.
* Oluşan __devtype_df__ DataFrame'ine ```Gender``` sütununu ekleyin.
* DataFrame'i cinsiyete göre gruplayarak her bir ```DevType``` için cinsiyete göre sayıları bulun.
* ```DevType``` tiplerindeki kadın-erkek oranlarını, ankete katılan kadın-erkek oranı ile karşılaştırarak kadınların daha çok tercih ettiği/edildiği ```DevType``` tiplerini listeleyin.

````python
#Her bir DevType tipinin sütun olarak, ankete katılanların da satır olarak temsil edildiği 
#yeni bir DataFrame oluşturup devtype_df adlı değişkene atayın.

arr4=survey_19['DevType'].str.split(';',expand=True).stack().unique()
devtype_df=pd.DataFrame(columns=arr4, index=survey_19.index)


erkek_say= (survey_19['Gender']=='Man').sum()
kadin_say= (survey_19['Gender']=='Woman').sum()
devtype_df=devtype_df.apply(lambda x:survey_19['DevType'].loc[x.index].str.contains(x.name))

#Oluşan devtype_df DataFrame'ine Gender sütununu ekleyin.
devtype_df['Gender'] = survey_19['Gender']

#DataFrame'i cinsiyete göre gruplayarak her bir DevType için cinsiyete göre sayıları bulun.
df4 = devtype_df.groupby('Gender').sum().T
df4['Ratio'] = df4['Woman']/df4['Man']> kadin_say/erkek_say

#DevType tiplerindeki kadın-erkek oranlarını, ankete katılan kadın-erkek oranı ile karşılaştırarak
#kadınların daha çok tercih ettiği/edildiği DevType tiplerini listeleyin.
print("Kadınların daha çok tercih ettiği DevType tipleri: ",list(df4[df4['Ratio']==True].index))
````

### Soru 9:
* ```MainBranch``` sütununda bulunan değerleri aşağıdaki değerler ile değiştirin.

    ``` {'I am a student who is learning to code': 'Student',
    'I am not primarily a developer, but I write code sometimes as part of my work': 'Not a Developer',
    'I am a developer by profession':  'Developer',
    'I code primarily as a hobby':   'Code as hobby',
    'I used to be a developer by profession, but no longer am': 'Was a Developer'} 
    ```
* Ana branş ```(MainBranch)``` X-ekseninde olacak şekilde her bir branştakilerin sayısını bar plot ile gösteriniz. Her bir bar, Cinsiyet ```(Gender)``` sayısına göre yığıt (stacked) şeklinde gösterilmelidir. 
* Ana branşı ```Developer``` olan ve Türkiye'den ankete katılanların eğitim seviyelerine göre dağılımını pasta grafiği ile gösteriniz.

````python
#MainBranch sütununda bulunan değerleri aşağıdaki değerler ile değiştirin.
# {'I am a student who is learning to code': 'Student',
#   'I am not primarily a developer, but I write code sometimes as part of my work': 'Not a Developer',
#   'I am a developer by profession':  'Developer',
#   'I code primarily as a hobby':   'Code as hobby',
#   'I used to be a developer by profession, but no longer am': 'Was a Developer'}
survey_19['MainBranch'].replace({'I am a student who is learning to code': 'Student',
  'I am not primarily a developer, but I write code sometimes as part of my work': 'Not a Developer',
  'I am a developer by profession':  'Developer',
  'I code primarily as a hobby':   'Code as hobby',
  'I used to be a developer by profession, but no longer am': 'Was a Developer'}, inplace=True)

#Ana branş (MainBranch) X-ekseninde olacak şekilde her bir branştakilerin sayısını bar plot ile gösteriniz.
#Her bir bar, Cinsiyet (Gender) sayısına göre yığıt (stacked) şeklinde gösterilmelidir.
df2 = survey_19.groupby(['MainBranch', 'Gender'])['MainBranch'].count().unstack()
df2[['Man','Non-binary', 'Woman']].plot(kind='bar', stacked=True)

#Ana branşı Developer olan ve Türkiye'den ankete katılanların eğitim seviyelerine göre dağılımını pasta grafiği ile gösteriniz.
data = survey_19['EdLevel'].loc[(survey_19['MainBranch']=='Developer')&(survey_19['Country']=='Turkey')]
counts = list(data.value_counts())
arr = list(data.value_counts().index)
plt.figure(figsize=(20,10))
plt.subplot(1,1,1)
plt.pie(counts,labels=arr,autopct='%1.2f%%')
plt.show()
````

### Soru 10:
Türkiye'den ankete katılan ve TL cinsinden maaş alanların ortalama yıllık brüt maaşını bulunuz.

### Soru 13:
* https://download.geonames.org/export/dump/countryInfo.txt adresindeki veriyi:
    * ilk 49 satırını ihmal ederek
    * yalnızca ```Country, Continent, CurrencyCode``` sütunlarını kullanarak 
    * ```Country``` sütununu indeks sütun olarak 
    * ```NA``` filtreleme yapmadan (```North America``` kıtasının kısaltması olan ```NA```'yı ```NaN``` değer olarak okumamak için)
  
  okuyarak __country_info__ adlı DataFrame'e aktarınız.
  
  
*     

           {AF : Africa
            AS : Asia			
            EU : Europe			
            NA : North America		
            OC : Oceania			
            SA : South America		
            AN : Antarctica	} 

   yukarıdaki haritalamayı kullanarak kıta isimlerini içeren ```ContinentName``` isimli yeni bir sütun oluşturunuz.

````python
# https://download.geonames.org/export/dump/countryInfo.txt adresindeki veriyi:
# ilk 49 satırını ihmal ederek
# yalnızca Country, Continent, CurrencyCode sütunlarını kullanarak
# Country sütununu indeks sütun olarak
# NA filtreleme yapmadan (North America kıtasının kısaltması olan NA'yı NaN değer olarak okumamak için)
# okuyarak country_info adlı DataFrame'e aktarınız.
country_info = pd.read_table(r"https://download.geonames.org/export/dump/countryInfo.txt", skiprows=49,
                             usecols=['Country','Continent','CurrencyCode'], keep_default_na=False)
country_info.set_index('Country', inplace=True)
# {AF : Africa
#   AS : Asia            
#   EU : Europe            
#   NA : North America        
#   OC : Oceania            
#   SA : South America        
#   AN : Antarctica    } 
# yukarıdaki haritalamayı kullanarak kıta isimlerini içeren ContinentName isimli yeni bir sütun oluşturunuz.
ContinentName = {'AF' : 'Africa','AS' : 'Asia','EU' : 'Europe','NA' : 'North America'
                 ,'OC' : 'Oceania','SA' : 'South America','AN' : 'Antarctica'}
country_info['ContinentName'] = country_info['Continent'].map(ContinentName)
````

### Soru 14:

* __survey_19__ dataframe'i ile __country_info__ dataframe'ini ülke bazında birleştirip __merged_survey__ isimli yeni bir dataframe'e aktarınız.
* Yaşadığı ülkenin resmi para biriminin dışındaki para birimiyle maaş alan kişi sayısı nedir?
* Her bir para biriminin kaç ülkede maaş olarak verildiğini azalan sırada listeleyiniz.
* Her kıtadaki ortalama maaşı en yüksek olan ülkeyi ve ortalama maaşı listeleyiniz.
* Avrupa kıtasındaki ülkelerin ortalama haftalık çalışma saati ```(WorkWeekHrs)``` ve ortalama yıllık maaşlarını ```(ConvertedComp)``` saçılım (scatter) grafiğinde gösteriniz.

````python
#survey_19 dataframe'i ile country_info dataframe'ini ülke bazında birleştirip merged_survey isimli
#yeni bir dataframe'e aktarınız.
merged_survey=survey_19.merge(country_info,on='Country')

#Yaşadığı ülkenin resmi para biriminin dışındaki para birimiyle maaş alan kişi sayısı nedir?
a=merged_survey[['CurrencySymbol','CurrencyCode']]
print("Yaşadığı ülkenin resmi para biriminin dışındaki para birimiyle maaş alan kişi sayısı:",
      a[a['CurrencySymbol'] != a['CurrencyCode']].shape[0])
````

````python
#Her bir para biriminin kaç ülkede maaş olarak verildiğini azalan sırada listeleyiniz.
test_df=merged_survey.groupby('CurrencySymbol')['Country'].value_counts().unstack().fillna(0)
test_df = test_df.apply(lambda x:  x.values>0)
test_df.sum(axis=1).sort_values(ascending=False)
````

````python
test_df = merged_survey[['ContinentName','Country','ConvertedComp']]
````

````python
#Her kıtadaki ortalama maaşı en yüksek olan ülkeyi ve ortalama maaşı listeleyiniz.
b =test_df.groupby(['ContinentName','Country']).mean().unstack()
b1=b.idxmax(axis=1).values
b2=b.max(axis=1).values
bdf = pd.DataFrame(data=b2, index = b1)
bdf.rename(columns= {0:'Ortalama'}, inplace = True)
bdf
````

````python
#Avrupa kıtasındaki ülkelerin ortalama haftalık çalışma saati (WorkWeekHrs) ve ortalama yıllık maaşlarını (ConvertedComp)
#saçılım (scatter) grafiğinde gösteriniz.
adad=merged_survey[merged_survey['ContinentName']=='Europe'].groupby('Country')['WorkWeekHrs','ConvertedComp'].mean()
axx = adad.plot(kind='scatter', x='WorkWeekHrs', y='ConvertedComp', figsize=(20,10))
for i, txt in enumerate(adad.index):
    axx.annotate(txt, (adad['WorkWeekHrs'][i], adad['ConvertedComp'][i]), ha='center', va='center')
````
